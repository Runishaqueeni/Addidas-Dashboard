```python
def generate_questions(user_input, top_n=10):
    print("NEW FUNCTION RUNNING")

    filtered_data = data[
        data["Skill"].str.contains(user_input, case=False, na=False)
    ]

    if filtered_data.empty:
        return []

    #OUTSIDE the if block
    filtered_text = (
        filtered_data["Role"] + " " +
        filtered_data["Skill"] + " " +
        filtered_data["Question"]
    )

    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.metrics.pairwise import cosine_similarity

    vectorizer = TfidfVectorizer()
    filtered_vectors = vectorizer.fit_transform(filtered_text)

    user_vec = vectorizer.transform([user_input])

    similarity = cosine_similarity(user_vec, filtered_vectors)
    top_indexes = similarity[0].argsort()[-top_n:][::-1]

    questions = []
    for i in top_indexes:
        questions.append(filtered_data.iloc[i]["Question"])

    return questions
```

```python
#skill = input("Enter skill:")
qs = generate_questions("Excel")

#print(qs)

for q in qs:
   print(q)
```
